```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,10000,30)
m = 9
C = 0.1
y_lin = m*x
y_quad = C*x*x
y_xlogx = np.log(x)*x
```

```python
plt.figure(figsize=(8,6),dpi=120)
plt.plot(x,y_lin,label="${:.2f}x$".format(m))
plt.plot(x,y_xlogx,label="$x \logx$")
plt.plot(x,y_quad,label="${:.2f}x^2$".format(C))
plt.legend()
plt.title("Growth rate comparison")
plt.xlabel("x")
plt.show()
```

```python
plt.figure(figsize=(8,6),dpi=120)
plt.plot(x,y_lin,label="${:.2f}x$".format(m))
plt.plot(x,y_xlogx,label="$x \logx$")
plt.legend()
plt.title("Growth rate comparison")
plt.xlabel("x")
plt.show()
```

Theorem: There does not exist a real number $C$ such that $Cx > x \log(x)$ for all sufficiently large $x$.
